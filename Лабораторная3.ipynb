**Простейшая модель Мальтуса роста населения Земли: $$y'= -ky$$**

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(y, t):
    k=-0.6
    return -k*y
```

```python
h=1/100
t=np.arange(0,10+h,h)

solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
```

**Метод Эйлера**

```python
def Euler(f, y0, t):
    y=[0]*len(t)
    y[0]=1
    
    for i in range(1,len(t)):
        y[i]=y[i-1]+h*f(y[i-1],t[i-1])
    return y 
```

```python
%matplotlib inline

plt.plot(t, Euler(dydt,1,t))
plt.plot(t, solve, label = r"$y'=-ky$",  color="green")
plt.grid(True)
plt.legend()
plt.show()
```
