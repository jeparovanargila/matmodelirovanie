$y'=k(y)y$.



```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    k=0.6
    return k*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=-ky$",  color="yellow")
plt.grid(True)
plt.legend()
plt.show()
```


$y'=k(y)y$.

$k(y)=a-by$.

$y'=a x -b\mathrm{x}^{2} $.

```python
def k(y):
    a=0.5
    b=0.2
    return  a-b*y
```

```python
def dydt(y, t):
    
    return k(y)*y
```

```python
t=np.linspace(0,10,100)
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=k(y)*y$")
plt.grid(True)
plt.legend()
plt.show()
```

$ y'=y - y^2 - c$

```python
def dydt(y, t):
    c=-0.4
    return y-y*y-c
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=y-y^2-c$", color="purple")
plt.grid(True)
plt.legend()
plt.show()
```


$y'= y - g(y)y - c$, где $g(y) = a - by$

```python
def g(y):
    a=-0.7
    b=-0.4
    return  a-b*y
```

```python
def dydt(y, t):
    с=-0.5
    return y-g(y)*y-с
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=y-g(y)y-c$", color="red")
plt.grid(True)
plt.legend()
plt.show()
```


$y'= y - y^2 - ky$

```python
def dydt(y, t):
    k=-0.5
    return y-y*y-y*k
```

```python
solve=odeint(dydt,y0=1,t=t)
solve=solve.flatten()
%matplotlib inline
plt.plot(t, solve, label = r"$y'=y-y^2-ky$", color="green")
plt.grid(True)
plt.legend()
plt.show()
```
